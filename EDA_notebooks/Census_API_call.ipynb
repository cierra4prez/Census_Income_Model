```python
import pandas as pd
import numpy as np
import requests 
import json
```

# API Call

```python
url = 'https://api.census.gov/data/2019/acs/acs1/pums?get=PINCP,WAGP,INTP,PWGTP,OCCP,NAICSP,WKHP,AGEP,SEX,FER,CPLT,PARTNER,COW,RAC1P,POBP,CIT,SCHL,WAOB'
response = requests.get(url)
response.status_code
```

```python
census = response.text
census = json.loads(census)
```

```python
df = pd.DataFrame.from_dict(census)
df
```

```python
# Set Header
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
df.head()
```
