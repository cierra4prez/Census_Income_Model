"Each housing and person record is assigned a weight,
because the records in the PUMS files represent a sample of the population. The weight is a numeric variable
expressing the number of housing units or people that
an individual microdata record represents. The sum of
the housing unit and person weights for a geographic
area is equal to the estimate of the total number of
housing units and people in that area. To generate
estimates based on the PUMS records, data users must
correctly apply weights."

Variables requiring ADJINC on the Person
files are: INTP, OIP, PAP, PERNP, PINCP, RETP, SEMP, SSIP, SSP, and WAGP.

```python
#API_KEY = ""
#API_KEY = "/Users/cierra4prez/.secret/census_api.json"

url = 'https://api.census.gov/data/2019/acs/acs1/pums?get=PINCP,WAGP,INTP,ADJINC,PWGTP,OCCP,NAICSP,WKHP,AGEP,SEX,FER,CPLT,PARTNER,COW,RAC1P,POBP,CIT,SCHL,DIS'
```

```python
response = requests.get(url)
response.status_code
```

```python
census = response.text
census = json.loads(census)
```

```python
df = pd.DataFrame.from_dict(census)
df
```
